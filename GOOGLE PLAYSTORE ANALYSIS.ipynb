### top apps with most download

```python
dx.groupby('Category').sum().sort_values(by='Installs',ascending=False)

top=dx.groupby('Category').sum().sort_values(by='Installs',ascending=False).head(10)
top
```

### what category are profitable and market leaders

```python
data['Price'].info()
```

```python
data['Price']=data['Price'].astype('float64').astype('int64')
data['Price']
```

```python
app_prices=data[['Category','Installs','Price']]
app_prices
```

```python
app_prices.groupby('Category').sum().sort_values(by='Price',ascending=False)
```

```python
app_prices.groupby('Category').sum().sort_values(by=['Installs','Price'],ascending=[False,False])
```

```python
d2=data[['App','Reviews','Price','Installs','Rating']].sort_values(by='Installs',ascending=False).head(20)
d2
```

```python
d3=d2[d2['Rating']>=4.2].sort_values(by=['Installs','Rating'],ascending=[True,False])
d3
```

```python
app_review=d3[['App','Reviews','Rating']].head(5)
app_review
```

```python
plt.figure(figsize=(12,10))

plt.xlabel('APPS')
plt.ylabel('Reviews')
plt.title('APP REVIEWS')

plt.bar(app_review['App'],app_review['Reviews'],color='green')
plt.legend()
```

```python
plt.figure(figsize=(12,10))

plt.bar(app_review['App'],app_review['Rating'],color='blue',width=0.4)
plt.xlabel('APPS')
plt.ylabel('RATINGS')
plt.title('TOP APP RATINGS')

```
